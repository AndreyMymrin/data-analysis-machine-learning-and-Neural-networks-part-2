# Ориентарованные ацикличные графы

Пусть G = (V,E) - граф, (V - признаки, E - ребра)

V = $(X_1, ... , X_d)$ - множество случайных величин(признаков)

Опр. Распределение P набора слуайных величин сл. вел. V c плотностью p(V) называется **Марковским** для графа G, если $p(V) = \prod^d_{j=1}p(X_j|\pi_j)$ где $X_j$ - значение в вершине, $\pi_j$ - непосредственно родители вершины $X_j$

M(G) - мн-во всех марковских распределений

# Оценить распределение на графе

Пусть G = (V,E); $V = (X_1, ..., X_d)$; $P_\theta из M(G)$ с плотностью $p_\theta

$\theta - ?$

Пусть $V_1, ..., V_n$ - данные(независимые реализации)

Функция правдоподобия: $Z(\theta) = \prod^n_{i=1}\prod^d_{j=1}P_{\theta_j}(X_{ij}|\pi_{ij})$

где $\theta_j$ - параметр для j-й вершины; $X_{ij}$ - значение в j-й вершине для i-го объекта; $P_{\theta_j}(X_{ij}|\pi_{ij})$ - какая-то модель

если категориальная модель - таблица условных вероятностей

если -//- веществ - линейная регрессия $X_j $~ $\pi_j$

Интервенция(вмешательство) - подход к оценке причинно-след. эффекта путём воздействия на граф
