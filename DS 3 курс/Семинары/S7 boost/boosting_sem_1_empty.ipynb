```python
import numpy as np
import pandas as pd

from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import mean_squared_error as mse
from sklearn.tree import plot_tree

import scipy

sns.set(font_scale=1.8, palette='Set2')
```

# Простая часть

Сгенерируем данные.

```python
np.random.seed(42)
X_train = scipy.stats.uniform.rvs(loc=-3, scale = 6, size=(500, 1))
y_train = np.ravel(np.sin(X_train)) + scipy.stats.norm.rvs(scale=0.1, size=500)

plt.figure(figsize=(10, 8))
plt.scatter(X_train, y_train)

plt.show()
```

```python
X_train.shape, y_train.shape
```

Обучите градиентный бустинг `GradientBoostingRegressor` с параметром `learning_rate=1` и 10 деревьями.

```python
regressor = GradientBoostingRegressor(learning_rate=1, n_estimators=10)
```

Выведите параметры бустинга. Какие деревья используются в бустинге по умолчанию? 

```python
regressor.get_params()
```

**Ответ**

неглубокие

Нарисуйте первые 5 деревьев бустинга. Какие они?

```python
regressor.fit(X_train,y_train)
```

```python
regressor.estimators_.shape
```

```python
for elem in regressor.estimators_[:5]:
    plt.figure()
    plot_tree(elem[0])
```

**Ответ**

Неглубокие

У бустинга есть метод `staged_predict`. Он выдает предсказания композиции первых $i$ моделей бустинга. Используя этот метод, нарисуйте для каждого  $i$ два графика:
* График предсказаний первых $i$ моделей и истинных ответов;
* График ошибки на шаге $i$ и предсказаний $i$-й компоненты бустинга.

```python
staged = list(regressor.staged_predict(X_train))
```

```python
len(staged)
```

```python
for i in range(10):
    if i == 0:
      previous_predict = 0
    else:
      previous_predict = staged[i-1]

    error = mse(y_train, staged[i])
    print(f"MSE_{i} = {error}")
    current_step_predict = staged[i]
    
    # построение графиков
    plt.figure(figsize = (15,7))
    plt.subplot(121)
    plt.title(f"i = {i}")
    plt.scatter(X_train, y_train, label = "true")
    plt.scatter(X_train, current_step_predict, label = "prediction")
    plt.legend()
    
    plt.subplot(122)
    plt.title(f"i = {i}")
    plt.scatter(X_train, current_step_predict - y_train, label = "- y_train")
    plt.scatter(X_train, current_step_predict - previous_predict, label = "- prev")
    plt.legend()
```

Сделайте вывод. Как выглядят графики на разных шагах? Как обучается бустинг? 

**Вывод**

predict становится всё лучше и лучше, ошибка уменьшается

Видно, что каждое последующее дерево очень похоже на предыдущее, за исключением некоторых участков, где идёт исправление ошибок предыдущей модели, что согласуется с логикой построения бустинг-модели
