```python
accuracy_val_forest = []
accuracy_train_forest = []
for i in np.arange(1,401,4):
    forest = RandomForestClassifier(n_estimators=i)
    forest.fit(X_train, y_train)
    accuracy_val_forest.append(accuracy_score(y_val, forest.predict(X_val)))
    accuracy_train_forest.append(accuracy_score(y_train, forest.predict(X_train)))
```

```python
accuracy_val_boost = []
accuracy_train_boost = []
for i in range(400):
    accuracy_val_boost.append(accuracy_score(y_val, staged_val[i]))
    accuracy_train_boost.append(accuracy_score(y_train, staged_train[i]))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.arange(1,401,1), accuracy_val_boost, label = "boost, validation")
plt.plot(np.arange(1,401,1), accuracy_train_boost, label = "boost, train")
plt.plot(np.arange(1,401,4), accuracy_val_forest, label = "forest, validation")
plt.plot(np.arange(1,401,4), accuracy_train_forest, label = "forest, train")
plt.legend()
plt.legend()
plt.xlabel("n_estimators")
plt.ylabel("accuracy score");
```

Видно, что при n > 100 точность не особо растёт, но время обучения рости должно, поэтому будем использовать n_estimators = 100, а не 400

```python
accuracy_val_boost = []
accuracy_train_boost = []
accuracy_val_forest = []
accuracy_train_forest = []
for i in np.arange(3,16,1):
    boost = GradientBoostingClassifier(n_estimators=100, max_depth= i)
    forest = RandomForestClassifier(n_estimators=100, max_depth= i)
    boost.fit(X_train, y_train)
    forest.fit(X_train, y_train)
    accuracy_val_boost.append(accuracy_score(y_val, boost.predict(X_val)))
    accuracy_train_boost.append(accuracy_score(y_train, boost.predict(X_train)))
    accuracy_val_forest.append(accuracy_score(y_val, forest.predict(X_val)))
    accuracy_train_forest.append(accuracy_score(y_train, forest.predict(X_train)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.arange(3,16,1), accuracy_val_boost, label = "boost, validation")
plt.plot(np.arange(3,16,1), accuracy_train_boost, label = "boost, train")
plt.plot(np.arange(3,16,1), accuracy_val_forest, label = "forest, validation")
plt.plot(np.arange(3,16,1), accuracy_train_forest, label = "forest, train")
plt.legend()
plt.xlabel("max_depth")
plt.ylabel("accuracy score");
```

Лес: нам нужно переобучение деревьев, чтобы они были максимально разные, поэтому должен быть ~возрастающий график, что и наблюдается

```python
accuracy_val_boost = []
accuracy_train_boost = []
for i in np.linspace(0.05, 0.25, 100):
    boost = GradientBoostingClassifier(n_estimators=100, max_depth= 4, learning_rate=i)
    boost.fit(X_train, y_train)
    accuracy_val_boost.append(accuracy_score(y_val, boost.predict(X_val)))
    accuracy_train_boost.append(accuracy_score(y_train, boost.predict(X_train)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.linspace(0.05, 0.25, 100), accuracy_val_boost, label = "validation")
plt.plot(np.linspace(0.05, 0.25, 100), accuracy_train_boost, label = "train")
plt.legend()
plt.xlabel("learning rate")
plt.ylabel("accuracy score");
```

```python
X_train_alpha =pd.DataFrame( X_train_big["fAlpha"])
X_test_alpha = pd.DataFrame(X_test["fAlpha"])
```

```python
plt.scatter(X_train_alpha, y_train_big)
```

Если бы классы были разделены, графики ниже получились бы нормальными, но мы имеем какую-то функцию дирихле

```python
boost = GradientBoostingClassifier(n_estimators= 10, learning_rate= 0.1)
boost.fit(X_train_alpha, y_train_big)
```

```python
staged = list(boost.staged_predict(X_train_alpha))
staged_test = list(boost.staged_predict(X_test_alpha))
for i in range(10):
    if i == 0:
      previous_predict = 0
    else:
      previous_predict = staged[i-1]

    current_step_predict = staged[i]
    print(f"accuracy №{i}  test: {accuracy_score(y_test, staged_test[i]).round(4)},"\
          f"train: {accuracy_score(y_train_big, staged[i]).round(4)}")
    # построение графиков
    plt.figure(figsize = (20,7))
    plt.subplot(121)
    plt.title(f"i = {i}")

    plt.scatter(X_train_alpha, y_train_big, label = "true", alpha = 0.5, linewidths=0.05)
    
    plt.scatter(X_train_alpha, current_step_predict, label = "prediction", alpha = 0.5, linewidths=0.05)
    plt.legend()
    
    plt.subplot(122)
    plt.title(f"i = {i}")
    plt.scatter(X_train_alpha, current_step_predict - y_train_big, label = "- y_train",\
                alpha = 0.5,linewidths=0.05,)
    plt.scatter(X_train_alpha, current_step_predict - previous_predict, label = "- prev",\
                alpha = 0.5,linewidths=0.05,)
    plt.legend()
```

Судя по графикам, accuracy должна быть 0, но в реальности всё нормально. Это связано с кривым отображением точек, лежащих близко друг к другу

#### Загрузка и предобработка данных:

**Физика:**

```python
data = pd.read_csv('hessi.solar.flare.UP_To_2018.csv')
```

Уберем категориальные признаки, а также дату и время начала и время конца вспышки:

```python
data = data.drop(['peak','flare', 'start.date', 'start.time', 'end', 'flag.1', 'flag.2', 'flag.3', 'flag.4', 'flag.5'], axis=1)
```

```python
data["energy.kev"].unique()
```

```python
data = data.replace({'energy.kev': {'3-6': 0, '6-12': 1, '12-25': 2, '25-50': 3, '50-100': 4,\
                                    '100-300':5, '300-800':6, '800-7000':7, '7000-20000':8}})
X = data.drop('duration.s', axis=1)
y = data['duration.s']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Исследуем значение метрики от n_estimators

```python
XGB = XGBRegressor(n_estimators=200, random_state=42, silent=True)
Cat = CatBoostRegressor(n_estimators=200, random_state=42,silent=True)
LGBM = LGBMRegressor(n_estimators=200, random_state=42)
boost = GradientBoostingRegressor(n_estimators=200, random_state=42)
```

```python
XGB.fit(X_train, y_train)
Cat.fit(X_train, y_train)
LGBM.fit(X_train, y_train)
boost.fit(X_train, y_train)
```

```python
#s_XGB = list(XGB.get_booster().staged_predict(X_test))
pr_Cat = list(Cat.staged_predict(X_test))
#s_LGBM = list(LGBM.staged_predict(X_test))
pr_boost = list(boost.staged_predict(X_test))
```

```python
mape_boost = []
mape_Cat = []
for i in range(200):
    mape_Cat.append(MAPE(y_test, pr_Cat[i]))
    mape_boost.append(MAPE(y_test, pr_boost[i]))
```

```python
mape_xgb = []
mape_lgbm = []
for n in np.arange(1,201,4):
    XGB = XGBRegressor(n_estimators=n, random_state=42, silent=True)
    LGBM = LGBMRegressor(n_estimators=n, random_state=42)
    XGB.fit(X_train, y_train)
    LGBM.fit(X_train, y_train)
    mape_xgb.append(MAPE(y_test, XGB.predict(X_test)))
    mape_lgbm.append(MAPE(y_test, LGBM.predict(X_test)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.arange(1,201,1), mape_boost, label = "boost")
plt.plot(np.arange(1,201,1), mape_Cat, label = "Cat")
plt.plot(np.arange(1,201,4), mape_lgbm, label = "LGBM")
plt.plot(np.arange(1,201,4), mape_xgb, label = "XGB")
plt.legend()
plt.legend()
plt.xlabel("n_estimators")
plt.ylabel("MAPE");
```

XGB требует меньшее количество деревьев для обучения, чем остальные модели. Обычный бустинг даёт результаты заметно хуже, чем стальные модели при том же количестве деревьев.

```python
X_tr, y_tr = X_train[:5000], y_train[:5000]
```

```python
mape_xgb = []
mape_lgbm = []
mape_boost = []
mape_Cat = []
for depth in np.arange(1,14,1):
    XGB = XGBRegressor(n_estimators=100, max_depth = depth, silent=True)
    LGBM = LGBMRegressor(n_estimators=100,max_depth=depth, silent=True)
    Cat = CatBoostRegressor(n_estimators=100, max_depth= depth, silent=True)
    boost = GradientBoostingRegressor(n_estimators=100, max_depth=depth)
    
    XGB.fit(X_tr, y_tr)
    LGBM.fit(X_tr, y_tr)
    Cat.fit(X_tr, y_tr)
    boost.fit(X_tr, y_tr)
    
    mape_xgb.append(MAPE(y_test, XGB.predict(X_test)))
    mape_lgbm.append(MAPE(y_test, LGBM.predict(X_test)))
    mape_Cat.append(MAPE(y_test, Cat.predict(X_test)))
    mape_boost.append(MAPE(y_test, boost.predict(X_test)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.arange(1,14,1), mape_boost, label = "boost")
plt.plot(np.arange(1,14,1), mape_Cat, label = "Cat")
plt.plot(np.arange(1,14,1), mape_lgbm, label = "LGBM")
plt.plot(np.arange(1,14,1), mape_xgb, label = "XGB")
plt.legend()
plt.legend()
plt.xlabel("max depth")
plt.ylabel("MAPE");
```

При высокой глубине деревьев для boost, Cat, xgb наблюдается переобучение, но для lgbm - равномерный спуск

```python
mape_xgb = []
mape_lgbm = []
mape_boost = []
mape_Cat = []
for rate in np.linspace(0.05,0.5, 40):
    XGB = XGBRegressor(n_estimators=100, max_depth = 5,learning_rate=rate , silent=True)
    LGBM = LGBMRegressor(n_estimators=100,max_depth=5,learning_rate=rate, silent=True)
    Cat = CatBoostRegressor(n_estimators=100, max_depth= 5,learning_rate=rate, silent=True)
    boost = GradientBoostingRegressor(n_estimators=100, max_depth=5, learning_rate=rate)
    
    XGB.fit(X_tr, y_tr)
    LGBM.fit(X_tr, y_tr)
    Cat.fit(X_tr, y_tr)
    boost.fit(X_tr, y_tr)
    
    mape_xgb.append(MAPE(y_test, XGB.predict(X_test)))
    mape_lgbm.append(MAPE(y_test, LGBM.predict(X_test)))
    mape_Cat.append(MAPE(y_test, Cat.predict(X_test)))
    mape_boost.append(MAPE(y_test, boost.predict(X_test)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.linspace(0.05,0.5, 40), mape_boost, label = "boost")
plt.plot(np.linspace(0.05,0.5, 40), mape_Cat, label = "Cat")
plt.plot(np.linspace(0.05,0.5, 40), mape_lgbm, label = "LGBM")
plt.plot(np.linspace(0.05,0.5, 40), mape_xgb, label = "XGB")
plt.legend()
plt.legend()
plt.xlabel("learning rate")
plt.ylabel("MAPE");
```

Ожидаемый рост погрешности при слишком большом lr. Но для Cat рост начинается заметно позже, чем у остальных.

```python
mape_xgb = []
mape_lgbm = []
mape_boost = []
mape_Cat = []
for alpha in np.linspace(0.005,1, 100):
    XGB = XGBRegressor(n_estimators=100, max_depth = 5,learning_rate=0.2, alpha = alpha, silent=True)
    LGBM = LGBMRegressor(n_estimators=100,max_depth=5,learning_rate=0.2,alpha = alpha, silent=True)
    boost = GradientBoostingRegressor(n_estimators=100, max_depth=5,alpha=alpha, learning_rate=0.2)
    
    XGB.fit(X_tr, y_tr)
    LGBM.fit(X_tr, y_tr)
    boost.fit(X_tr, y_tr)
    
    mape_xgb.append(MAPE(y_test, XGB.predict(X_test)))
    mape_lgbm.append(MAPE(y_test, LGBM.predict(X_test)))
    mape_boost.append(MAPE(y_test, boost.predict(X_test)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.linspace(0.05,1, 99), mape_boost, label = "boost")
plt.plot(np.linspace(0.05,1, 99), mape_lgbm, label = "LGBM")
plt.plot(np.linspace(0.05,1, 99), mape_xgb, label = "XGB")
plt.legend()
plt.legend()
plt.xlabel("L1 reg. parametr")
plt.ylabel("MAPE");
```

```python
mape_xgb = []
mape_lgbm = []
mape_boost = []
mape_Cat = []
for l2_param in np.linspace(0.005,1, 100):
    XGB = XGBRegressor(n_estimators=100, max_depth = 5,learning_rate=0.2,\
                       alpha = 0.1,l2_leaf_reg = l2_param, silent=True)
    
    LGBM = LGBMRegressor(n_estimators=100,max_depth=5,learning_rate=0.2,\
                         alpha = 0.1,l2_leaf_reg = l2_param, silent=True)
    
    Cat = CatBoostRegressor(n_estimators=100, max_depth= 5,learning_rate=0.2\
                            ,l2_leaf_reg= l2_param, silent=True)
    
    
    XGB.fit(X_tr, y_tr)
    LGBM.fit(X_tr, y_tr)
    Cat.fit(X_tr, y_tr)
    
    mape_xgb.append(MAPE(y_test, XGB.predict(X_test)))
    mape_lgbm.append(MAPE(y_test, LGBM.predict(X_test)))
    mape_Cat.append(MAPE(y_test, Cat.predict(X_test)))
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.linspace(0.005,1, 100), mape_Cat, label = "Cat")
plt.plot(np.linspace(0.005,1, 100), mape_lgbm, label = "LGBM")
plt.plot(np.linspace(0.005,1, 100), mape_xgb, label = "XGB")
plt.legend()
plt.legend()
plt.xlabel("L2 reg. parametr")
plt.ylabel("MAPE");
```

Странности:
1)  У LGBM есть параметры регуляризации, но они, похоже, ни на что не влияют (?!)

2) Boost на L1 и Cat на L2 ведут себя хаотчно.

3) параметр L2 регуляризации не влияет на точность для XGB

3) Cat предсказывает сильно быстрее остальных

4) MAPE примерно одинакова, поэтому главный плюс усложненный бустингов по сравнению с обычным - скорость фита
